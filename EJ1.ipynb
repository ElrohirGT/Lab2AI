# Ejercicio 1

Escribe un código en Python para simular lanzamientos de una moneda en la computadora. Se debe permitir que el usuario
elija parámetro 0 < p < 1 que indica la probabilidad de obtener un ´exito en el lanzamiento de la moneda. Hacer lo siguiente:

a) Obtener, mediante repeticiones, una estimación de la densidad del número de lanzamientos necesarios para obtener el
primer éxito. Por ejemplo, simular un experimento de estos N veces. Usar N = 1000.

b) Elaborar 3 visualizaciones de la función de densidad o masa, y cambiando los valores de p.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import geom
```

```python
# Definir parámetros
N = 1000
p_values = [0.2, 0.5, 0.8]

# Configurar visualización
plt.figure(figsize=(12, 4))

for i, p in enumerate(p_values, 1):
    # Simular N experimentos
    data = np.random.geometric(p, N)
    
    # Obtener frecuencias
    unique, counts = np.unique(data, return_counts=True)
    empirical_pmf = counts / N
    
    # Distribución teórica
    k = np.arange(1, max(data) + 1)
    theoretical_pmf = geom.pmf(k, p)
    
    # Graficar
    plt.subplot(1, 3, i)
    sns.barplot(x=unique, y=empirical_pmf, color="blue", label="Simulada")
    plt.plot(k, theoretical_pmf, "ro-", label="Teórica")
    
    plt.xlabel("Lanzamientos hasta el primer éxito")
    plt.ylabel("Probabilidad")
    plt.title(f"p = {p}")
    plt.legend()

plt.tight_layout()
plt.show()
```
