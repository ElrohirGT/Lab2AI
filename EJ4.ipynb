```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

n = 4
mu = np.array([0, 1, 10, 21])
Sigma = np.array([
    [1.0, 0.5, 0.2, 0.0],
    [0.5, 1.5, 3.8, 0.1],
    [0.2, 0.3, 1.0, 0.4],
    [0.0, 0.2, 0.4, 2.0]
])

mu, Sigma
```

```python

size = 1000
muestra = np.random.multivariate_normal(mu, Sigma, size=size)
muestra
```

```python

column_names = [f"X{i+1}" for i in range(n)]
df = pd.DataFrame(muestra, columns=column_names)
df
```

```python

sns.pairplot(df, diag_kind="hist", corner=False)
plt.suptitle("Pairplot de muestra multivariada", y=1.02)
plt.show()
```

```python

media_muestral = df.mean().values
covarianza_muestral = df.cov().values

print("Media teÃ³rica (mu):")
print(mu)
print("\nMedia muestral aproximada:")
print(media_muestral)

print("\nCovarianza teÃ³rica (Sigma):")
print(Sigma)
print("\nCovarianza muestral aproximada:")
print(covarianza_muestral)
```
