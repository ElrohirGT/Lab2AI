```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
file_path = "weather.csv"
df = pd.read_csv(file_path)
df
```

```python

X = df.iloc[:, :-1].values
stations = df["station"].values

pca = PCA(n_components=2)
principal_components = pca.fit_transform(X)
principal_components
```

```python
plt.figure(figsize=(10, 5))
for i in range(len(principal_components)):
    plt.plot(range(1, 13), principal_components[i, 0] * np.ones(12), label=stations[i] if i < 3 else "", alpha=0.5)

plt.xlabel("Mes")
plt.ylabel("Componente Principal 1")
plt.title("VariaciÃ³n de la Primera Componente Principal por Mes")
plt.show()

plt.figure(figsize=(10, 5))
for i in range(len(principal_components)):
    plt.plot(range(1, 13), principal_components[i, 1] * np.ones(12), label=stations[i] if i < 3 else "", alpha=0.5)

plt.xlabel("Mes")
plt.ylabel("Componente Principal 2")
plt.title("VariaciÃ³n de la Segunda Componente Principal por Mes")
plt.show()


```

```python
plt.figure(figsize=(10, 6))
plt.scatter(principal_components[:, 0], principal_components[:, 1], alpha=0.7)
for i, station in enumerate(stations):
    plt.text(principal_components[i, 0], principal_components[i, 1], station, fontsize=8, alpha=0.7)

plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.title("Biplot de las estaciones en el espacio de PCA")
plt.grid()
plt.show()
```
